```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/Toshiba/OneDrive/Desktop/cleandata.csv')

# Mapping for color scheme labels (Response scheme)
color_labels_response = {
    1: 'Yes',
    2: 'No',
    3: "Don't know",
    4: 'Refused to answer'
}

# Create a new DataFrame with gender and variables of interest
df_gender_vars = df[['Gender', 'fin11a', 'fin11b', 'fin11c', 'fin11d', 'fin11e', 'fin11f', 'fin11g', 'fin11h']].copy()

# Replace numerical codes with labels for the response scheme
df_gender_vars.replace(color_labels_response, inplace=True)

# Reshape the DataFrame for plotting
df_gender_vars = pd.melt(df_gender_vars, id_vars='Gender', var_name='Variable', value_name='Response')

# Set the size of the graph
plt.figure(figsize=(10, 6))

# Plotting the graph
ax = sns.countplot(x='Variable', hue='Response', data=df_gender_vars, hue_order=color_labels_response.values())

# Adding labels and title
plt.xlabel('Reason For No Account')
plt.ylabel('Count')
plt.title('Reasons For No Account Distribution')

# Move the legend outside the graph
ax.legend(title='Response', loc='center left', bbox_to_anchor=(1.025, 0.25))

# Create a legend box below the graph for reasons of no account as text
legend_text = '''Reasons for no account:
fin11a - Reason for no account: too far
fin11b - Reason for no account: too expensive
fin11c - Reason for no account: lack documentation
fin11d - Reason for no account: lack trust
fin11e - Reason for no account: religious reasons
fin11f - Reason for no account: lack money
fin11g - Reason for no account: family member already has one
fin11h - Reason for no account: no need for financial services'''

plt.text(1.025, 0.5, legend_text, transform=ax.transAxes, fontsize=10)

# Display the plot
plt.show()
```
