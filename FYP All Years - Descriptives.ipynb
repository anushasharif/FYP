```python
import pandas as pd

# Assuming you have a DataFrame called 'data' with columns: 'financial_inclusion', 'gender', 'age', 'employment', 'education', 'internet_access', 'owns_a_mobile', and 'year'
data = pd.read_excel('C:/Users/Toshiba/OneDrive/Desktop/alldata.xlsx')

# Group the data by 'year'
grouped_data = data.groupby('Year')

# Descriptive analysis for financial inclusion
financial_inclusion_stats = grouped_data['FinancialInc'].describe()

# Descriptive analysis for gender
gender_stats = grouped_data['Gender'].value_counts(normalize=True)

# Descriptive analysis for age
age_stats = grouped_data['age'].describe()

# Descriptive analysis for employment
employment_stats = grouped_data['emp_in'].value_counts(normalize=True)

# Descriptive analysis for education
education_stats = grouped_data['educ'].value_counts(normalize=True)

# Descriptive analysis for internet access
internet_access_stats = grouped_data['internetaccess'].value_counts(normalize=True)

# Descriptive analysis for owns a mobile
owns_a_mobile_stats = grouped_data['mobileowner'].value_counts(normalize=True)

# Print the results
print("Financial Inclusion:")
print(financial_inclusion_stats)
print("\nGender:")
print(gender_stats)
print("\nAge:")
print(age_stats)
print("\nEmployment:")
print(employment_stats)
print("\nEducation:")
print(education_stats)
print("\nInternet Access:")
print(internet_access_stats)
print("\nOwns a Mobile:")
print(owns_a_mobile_stats)
```
