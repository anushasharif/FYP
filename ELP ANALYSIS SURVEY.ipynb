```python
import pandas as pd

# Assuming you have a pandas DataFrame named 'survey_data' with the survey responses
# and the Yes/No variables are stored in columns named 'Q18', 'Q12', 'Q20', 'Q3', 'Q9', 'Q10', 'Q11'

# Load the survey data into a pandas DataFrame
data = pd.read_excel('C:/Users/Toshiba/OneDrive/Desktop/BISMACLEANDATA.xlsx')

# Select only the columns containing the specific Yes/No variables
yes_no_columns = ['Q18', 'Q12', 'Q20', 'Q3', 'Q9', 'Q10', 'Q11']
yes_no_data = data[yes_no_columns]

# Calculate descriptive statistics
statistics = yes_no_data.count()

# Display the statistics
print(statistics)
```

```python
variables = ['Q18', 'Q12', 'Q20', 'Q3', 'Q9', 'Q10', 'Q11']

# Count the occurrences of each value in each variable
for variable in variables:
    value_counts = data[variable].value_counts()
    print(f"Question: {variable}")
    print(value_counts)
    print()
```

```python
# Select the columns containing the variables
variables = ['Q18', 'Q12', 'Q20', 'Q3', 'Q9', 'Q10', 'Q11']

# Perform cross-tabulation for each pair of variables
for i in range(len(variables)):
    for j in range(i + 1, len(variables)):
        variable1 = variables[i]
        variable2 = variables[j]
        cross_tab = pd.crosstab(data[variable1], data[variable2])
        print(f"Cross-tabulation between {variable1} and {variable2}:")
        print(cross_tab)
        print()
```

```python
# Select the columns containing the variables
variables = ['Q18', 'Q12', 'Q20', 'Q3', 'Q9', 'Q10', 'Q11']

# Perform cross-tabulation for each pair of variables
for i in range(len(variables)):
    for j in range(i + 1, len(variables)):
        variable1 = variables[i]
        variable2 = variables[j]
        cross_tab = pd.crosstab(data[variable1], data[variable2], normalize='index') * 100
        print(f"Cross-tabulation in % between {variable1} and {variable2}:")
        print(cross_tab)
        print()
```
